###Filing NaN values with Mean
Problem:New value becomes high because of outlier(Large Value Scale)

```python
MeandatasetnotNaN=dataset.fillna(dataset.mean())
MeandatasetnotNaN
```

###Filing NaN values with Median

```python
MediandatasetnotNaN = dataset.fillna(dataset.price.median())
MediandatasetnotNaN
```

###Identifying and removing Outlier : Huge difference in Value

```python
dataset.describe()
```

```python
percentile = dataset.price.quantile(1)
percentile
```

```python
datasetnoOutlier = dataset[dataset.price<percentile]
datasetnoOutlier
```
